```python
from wordcloud import WordCloud, STOPWORDS
```

```python
stopwords = set (STOPWORDS)
```

```python
alice_wc = WordCloud (
    background_color = 'white',
    max_words = 2000,
    stopwords = stopwords
)

alice_wc.generate(alice_text)
```

```python
plt.imshow (alice_wc, interpolation = 'bilinear')
plt.axis ('off')
plt.show ()
```

```python
fig = plt.figure ()
fig.set_figwidth (14)
fig.set_figheight (18)

plt.imshow (alice_wc, interpolation = 'bilinear')
plt.axis ('off')
plt.show ()
```

```python
from PIL import Image

alice_mask = np.array (Image.open('alice_mask.png'))
```

```python
fig = plt.figure ()
fig.set_figwidth (14)
fig.set_figheight (18)

plt.imshow (alice_mask, cmap=plt.cm.gray, interpolation = 'bilinear')
plt.axis ('off')
plt.show ()
```

```python
alice_wc = WordCloud (
    background_color = 'white',
    max_words = 2000,
    mask = alice_mask,
    stopwords = stopwords
)

alice_wc.generate(alice_text)

fig = plt.figure ()
fig.set_figwidth (14)
fig.set_figheight (18)

plt.imshow (alice_wc, interpolation = 'bilinear')
plt.axis ('off')
plt.show ()
```
